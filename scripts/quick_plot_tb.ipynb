```python
# Simple Hash
df1 = pd.read_csv("/home/tibigg/Downloads/tables/d#_dqn.csv", usecols=["Step", "Value"])
df2 = pd.read_csv("/home/tibigg/Downloads/tables/d#_dqn_a.csv", usecols=["Step", "Value"])
df3 = pd.read_csv("/home/tibigg/Downloads/tables/d#_maddpg.csv", usecols=["Step","Value"])
df4 = pd.read_csv("/home/tibigg/Downloads/tables/d#_maddpg_a.csv", usecols=["Step","Value"])
```

```python
dfs = [(df1, "DQN"), (df2, "DQN + Ego Attention"), (df3, "MADDPG"), (df4, "MADDPG + Self Attention")]

import matplotlib.pyplot as plt

ts_factor = 0.97
for (df, label) in dfs:
    smooth = df.ewm(alpha=(1 - ts_factor)).mean()
    # plt.plot(df["Value"], alpha=0.1)
    plt.plot(smooth["Value"], label=label)

plt.title("Deadlocked # Total Reward/Episode")
plt.xlabel("Episode")
plt.ylabel("Total Reward")
plt.legend(loc='best')
plt.grid(alpha=0.3)

plt.savefig("/home/tibigg/Downloads/tables/d#.png")
plt.show()
```
